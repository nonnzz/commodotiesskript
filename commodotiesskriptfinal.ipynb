```python

with open("C:/Users/Nabdelsalam/OneDrive - Afreximbank/digital solutions/commo/Commodity.out", 'r') as r, open('C:/Users/Nabdelsalam/OneDrive - Afreximbank/digital solutions/commo/file.txt.txt', 'w') as o:
    for line in r:
        #strip() function
        if line.strip():
            o.write(line)
commodoties = open("C:/Users/Nabdelsalam/OneDrive - Afreximbank/digital solutions/commo/file.txt.txt", "r")
#print("New text file:\n",f.read())

temp=[]
fin=[]
for x in commodoties:
    temp.append(x)
# print(temp)

for j in temp:
    field=j.split()
    code= field[0]
    # print(code)
    if code=="CO1" or code == 'XAU' or code =='LMCODY' or code =='LA1' or	code =='HG1' or code =='LMZSDS03' or code =='CC1' or	code =='DF1' or 	code =='CT1'	or code == 'SB1'	or code =='W' or	code =='C' or	code == 'KO1' or	code == 'NG1':
            fin.append(j)
# print(fin[0])            
# print(len(fin))
arrs=[]
for x in fin:
     arr= x.split('|')
     arrs.append(arr)

# labels=['codes','date','price',"ran"]
df= pd.DataFrame(np.row_stack(arrs))
df=df.drop(df.iloc[:, 3:], axis=1)
df.set_axis(['code','date','price'], axis='columns', inplace=True)
# df=df.columns = []
# df.head()
```

```python
sheet= df.pivot_table(index='date', columns='code')
sheet.set_axis(['corn','cocoa','cotton','coffee','copper','palm oil','ali','cobalt','zinc','sugar','wheat','gold'], axis='columns', inplace=True)
sheet = sheet.replace(np.nan, None)
sheet.index = pd.to_datetime(sheet.index, format = '%d-%b-%y').strftime('%Y-%m-%d')
sheet.head(30)
```
